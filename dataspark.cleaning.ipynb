```python
import pandas as pd
import re
import numpy as np
```

```python
# Try reading the file with 'latin-1' encoding
df_customer = pd.read_csv("D:/MDTM27/capstone project/Data spark/Customers.csv", encoding='latin-1')
df_customer
```

```python
#alphapet value

import re
df_customer['Name'] = df_customer['Name'].apply(lambda x: re.sub('[^a-zA-Z]', '', x))

df_customer['Name']
```

```python
df_customer['Birthday'] = df_customer['Birthday'].str.replace("/","-")
df_customer['Birthday'].head()
```

```python
df_customer['Birthday'] = df_customer['Birthday'].apply(lambda x: '-'.join([i.zfill(2) for i in x.split('-')]))
df_customer['Birthday'].head()
```

```python
#Birthday convered to datetime functionality
df_customer['Birthday'] = pd.to_datetime(df_customer['Birthday'], infer_datetime_format=True, errors='coerce') 
df_customer['Birthday'].head()
```

```python
# zip code dtype convered to int
df_customer['Zip Code'] = pd.to_numeric(df_customer['Zip Code'], errors='coerce').fillna(0).astype(int)

df_customer
```

```python
df_customer['State Code'].fillna(method='ffill', inplace=True) 
df_customer
```

```python
#customer in csv file
df_customer.to_csv('customer_data.csv', index=False)
```

```python
#sales data
df_sales = pd.read_csv("D:/MDTM27/capstone project/Data spark/Sales.csv")
df_sales
```

```python
#/ are replace the '-'

df_sales['Order Date'] = df_sales['Order Date'].str.replace("/","-")
df_sales['Delivery Date'] = df_sales['Delivery Date'].str.replace("/","-")
df_sales[['Order Date','Delivery Date']].head()
```

```python
#Order Date,Delivery Date converted to the datatype
df_sales['Order Date'] = pd.to_datetime(df_sales['Order Date'], infer_datetime_format=True, errors='coerce') 
df_sales['Delivery Date'] = pd.to_datetime(df_sales['Delivery Date'], infer_datetime_format=True, errors='coerce') 
df_sales[['Order Date','Delivery Date']].head()
```

```python
#sales data store in csv file
df_sales .to_csv('sales_data.csv', index=False)
```

```python
df_products = pd.read_csv("D:/MDTM27/capstone project/Data spark/Products.csv")
df_products 
```

```python
# Convert all values to strings first
df_products['Unit Cost USD'] = df_products['Unit Cost USD'].astype(str)
df_products['Unit Price USD'] = df_products['Unit Price USD'].astype(str)

# Strip the $ remove the Unit Cost USD,Unit Price USD
df_products['Unit Cost USD'] = df_products['Unit Cost USD'].str.strip("$")
df_products['Unit Price USD'] = df_products['Unit Price USD'].str.strip("$")

# Convert the cleaned strings to numeric values Unit Cost USD,Unit Price USD
df_products['Unit Cost USD'] = pd.to_numeric(df_products['Unit Cost USD'], errors='coerce')
df_products['Unit Price USD'] = pd.to_numeric(df_products['Unit Price USD'], errors='coerce')
df_products[['Unit Cost USD','Unit Price USD']]
```

```python
#product data store in csv file
df_products .to_csv('product_data.csv', index=False)
```
